```python
import pandas as pd
```

```python
df = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
n_df = df.copy()
```

```python
n_df.info()
```

```python
n_df = n_df.drop(['User ID'], axis=1)
```

```python
n_df
```

```python
n_df_enco = pd.get_dummies(n_df)
n_df_enco
```

```python
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.metrics import r2_score, mean_squared_error
```

```python
train, test = train_test_split(n_df_enco, test_size=0.2, random_state=0)
```

```python
predictor = list((n_df_enco.drop('EstimatedSalary', axis=1)).columns)
target = ['EstimatedSalary']
```

#Using SatandardScaler

```python
scaler = StandardScaler()
train_predictor_scaled = scaler.fit_transform(train[predictor])
train_predictor_scaled = pd.DataFrame(train_predictor_scaled)

# test_predictor_scaled = scaler.fit_transform(test[predictor])
# test_predictor_scaled = pd.DataFrame(test_predictor_scaled)
```

```python
model = {'Linear Regression' : LinearRegression(),
         'Ridge' : Ridge(alpha=2),
         'Lasso' : Lasso(alpha=2),
         'ElasticNet' : ElasticNet(alpha=2)}

report = []

for name, obj in model.items():
  obj.fit(train_predictor_scaled, train[target])
  predicted_training = obj.predict(train_predictor_scaled)
  predicted_test = obj.predict(test_predictor_scaled)

  train_r2 = r2_score(train[target], predicted_training)
  train_mse = mean_squared_error(train[target], predicted_training)
  train_rmse = np.sqrt(mean_squared_error(train[target], predicted_training))

  test_r2 = r2_score(test[target], predicted_test)
  test_mse = mean_squared_error(test[target], predicted_test)
  test_rmse = np.sqrt(mean_squared_error(test[target], predicted_test))

  report.append([name, train_r2, train_rmse, train_mse, test_r2, test_rmse, test_mse])

report_df = pd.DataFrame(report, columns=['Model', 'train_r2', 'train_rmse', 'train_mse', 'test_r2', 'test_rmse', 'test_mse'])
```

```python
report_df
```
